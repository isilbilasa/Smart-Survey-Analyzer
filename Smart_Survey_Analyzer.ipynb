```python
import random
scores = []

while len(scores) < 20:
    score = random.randint(0,6)

    if score == 0:
        print("Anket kullanıcı tarafından sonlandırıldı")
        break

    if score == 6:
        continue

    scores.append(score)
```

```python
toplam_cevap = len(scores)
## While döngüsü kullandım çünkü döngü sayısının 'i' gibi bir aralık değil,
##'scores'listesinin uzunluğu veya 'score' değişkeninin değeri gibi koşullar belirliyor 
```

```python
if len(scores) > 0:
    ortalama = sum(scores) / len(scores)
else:
    print("Hesaplanacak veri bulunamadı.")
    
```

```python
en_dusuk_puan = min(scores)
en_yuksek_puan = max(scores)
```

```python
memnun = 0 
kararsiz = 0 
memnun_degil = 0

for score in scores:
    if score >= 4:
        memnun += 1
    elif score == 3:
        kararsiz += 1
    else:
        memnun_degil += 1
```

```python
arka_arkaya_bir_durumu = 0
for score in scores:
    if score == 1:
        arka_arkaya_bir_durumu += 1
    else:
        arka_arkaya_bir_durumu = 0
    if arka_arkaya_bir_durumu == 3:
        print("⚠️ Kritik memnuniyetsizlik tespit edildi: Üst üste 3 adet 1 puan!")
        break
```

```python
print(f"Toplam geçerli cevap: {len(scores)}")
print(f"Ortalama puan: {round(sum(scores)/len(scores),2) if len(scores) > 0 else 0}")
print(f"Min puan: {min(scores) if scores else 0}")
print(f"Max puan: {max(scores) if scores else 0}")
print("-" * 30)
print(f"Memnun: {memnun}")
print(f"Kararsız: {kararsiz}")
print(f"Memnun değil: {memnun_degil}")
```
